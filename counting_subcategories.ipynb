# Explode subcategories

```python
import pandas as pd
#load Pandas for data structures and operations for manipulating numerical tables and time series.
import numpy as np
#Load NumPy for linear algebra/ Numerical Python
```

```python
bern = (
    pd.read_csv('bern_matters_expanded.csv')
    .rename(str.lower, axis='columns')
)

# Met de volgende regel maken we van de waarde in het veld subcategories een array (list)
# door te splitsen op de '; ' die tussen elke subcategorie staat.
bern = bern.assign( matters=bern[ 'subcategories' ].str.split( ', ' ) )

# Met de functie 'explode' kunnen we nu voor elke subcategorie in die array
# een duplicaat-regel aanmaken.
bern = bern.explode( 'subcategories' )

# Bestudeer het effect in deze eerste 100 regels
bern.head( n=100 )
```

```python
# Merk op dat de tabel nu dus veel meer regels heeft!
# Voor elke subcategorie in een regel is er nu een duplicaat-regel.
# Je moet deze tabel dus NIET gebruiken voor bijvoorbeeld het tellen van 
# 'lawid', want dan krijg je er veel te veel.
bern
```

```python
# En nu kunnen we dus vrij simpel 'doorselecteren'
bern[ bern['secondarycategory']=='Juden' ]['subcategories'].value_counts()
```

```python
bern[ bern['secondarycategory']=='Zigeuner' ]['subcategories'].value_counts()
```

```python
# Set is 87 regels lang, dus voor display..
m = bern[ bern['secondarycategory']=='Fahrende Leute' ]['subcategories'].value_counts()
for idx in m.index:
    print( idx[0:50].ljust(54), '{:3d}'.format( m[idx] )  )
```
