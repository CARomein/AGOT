```python
def extract_id(x): 
    idnum = x[0]
    if idnum[0] == '(':
        idnum = idnum[1:]
    return idnum

def extract_category_num(x):
    return x[2]

def extract_main_categories(x): 
    main = x[3]
    secondary = None
    if x[4] == '': 
        secondary = x[5]
    return main, secondary

def extract_sub_categories(x): 
    splitted = matter.split(',,,,,')

    return splitted[1]
```

```python
li = []

for i, row in df.iterrows(): 
    if i != 1:
        continue

    new_row = row.copy()
    matters = row['Matters']
    
    if len(matters) == 0:
        print('this is the case')
    for matter in matters.split('),('):
        if matter == '':
            continue
            
        splitted = matter.split(',')
        
        idnum = extract_id(splitted)
        cat_num = extract_category_num(splitted)
        main, secondary = extract_main_categories(splitted)
        subcategories = extract_sub_categories(matter)
        
        print(matter)
        print(f'\nID: {idnum}')
        print(f'Category number: {cat_num}')
        print(f'Main category: {main} \t Secondary {secondary}')
        print(f'Subcategories: {subcategories} \n\n\n')
        
        
        li.append([row['LawID'], idnum, cat_num, main, secondary, subcategories])
```

```python
new_df = pd.DataFrame(li, columns = ['LawID', 'SubLawID', 'CategoryNumber', 'MainCategory', 'SecondaryCategory', 'SubCategories'])
new_df.to_csv('zuerich_test.csv')
```
